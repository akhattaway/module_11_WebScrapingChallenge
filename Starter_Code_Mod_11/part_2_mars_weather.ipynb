### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
mars_months = mars_df['month'].value_counts().sort_index(ascending=True)
mars_months
```

```python
# 2. How many Martian days' worth of data are there?
mars_days = mars_df['terrestrial_date'].nunique()
mars_days
```

```python
# 3. What is the average low temperature by month?
avg_monthly_temp = mars_df.groupby('month')['min_temp'].mean()
avg_monthly_temp
```

```python
# Plot the average temperature by month

# Group the DataFrame by 'month' and calculate the mean of 'min_temp'
avg_monthly_temp = mars_df.groupby('month')['min_temp'].mean()

# Plot the average temperature by month as a bar graph
avg_monthly_temp.plot(kind='bar', color='blue')

# Set the title and labels
plt.title('Average Temperature by Month on Mars')
plt.xlabel('Month')
plt.ylabel('Average Temperature (C)')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location

# Find the coldest temp
coldest_temp = mars_df['min_temp'].min()

# Filter the DataFrame for rows where 'min_temp' matches the coldest temperature
coldest_month_df = mars_df[mars_df['min_temp'] == coldest_temp]

# Find the coldest month
coldest_month = coldest_month_df['month'].iloc[0]

# Find the hottest temp
hottest_temp = mars_df['min_temp'].max()

# Filter the DataFrame for rows where 'min_temp' matches the hottest temperature
hottest_month_df = mars_df[mars_df['min_temp'] == hottest_temp]

# Extract the month from the filtered DataFrame
hottest_month = hottest_month_df['month'].iloc[0]


print(f"coldest month:",coldest_month,"coldest temp:", coldest_temp)
print(f"hottest month:",hottest_month,"hottest temp:", hottest_temp)
```

```python
# 4. Average pressure by Martian month
avg_monthly_pressure = mars_df.groupby('month')['pressure'].mean()
avg_monthly_pressure
```

```python
# Plot the average pressure by month

# Calculate the average pressure by month
avg_monthly_pressure = mars_df.groupby('month')['pressure'].mean()

# Plot the average pressure by month
plt.figure(figsize=(10, 6))
avg_monthly_pressure.plot(kind='bar', color='skyblue')
plt.title('Average Pressure by Month on Mars')
plt.xlabel('Month')
plt.ylabel('Average Pressure')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?

# Find the earliest and latest dates
earliest_date = mars_df['terrestrial_date'].min()
latest_date = mars_df['terrestrial_date'].max()

# Calculate the difference in days
days_covered = (latest_date - earliest_date).days
days_covered

```

```python
import numpy as np
# Plot temperatures as a line graph
plt.figure(figsize=(10, 6))
plt.plot(np.arange(len(mars_df)), mars_df['min_temp'], color='green', label='Temperatures')

# Customize the plot
plt.title('Temperatures on Mars Over Time')
plt.xlabel('Days')
plt.ylabel('Temperature (C)')
plt.legend()
plt.grid(True)

# Set custom x-axis ticks
tick_positions = np.arange(0, len(mars_df), 250)
tick_labels = np.arange(0, 2000, 250)
plt.xticks(tick_positions, tick_labels)

plt.tight_layout()

# Display the plot
plt.show()
```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
